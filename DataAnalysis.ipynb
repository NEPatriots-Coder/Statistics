```python
import pandas as pd
import numpy as np
import seaborn as sns                       #visualisation
import matplotlib.pyplot as plt             #visualisation
%matplotlib inline
sns.set(color_codes=True)
```

### Below I will calculate the Interquartile Range which will measure the statistical dispersion with which the middle 50% of the data falls.

```python
Q1 = df.quantile(0.25, numeric_only=True)
Q3 = df.quantile(0.75, numeric_only=True)
IQR = Q3 - Q1
print(IQR)
```

### Removing outliers

```python
Q1_copy = Q1.copy()
Q3_copy = Q3.copy()
```

```python
filtered_df = df[~((df < (Q1_copy - 1.5 * IQR))
| (df > (Q3_copy + 1.5 * IQR))).any(axis=1)]

print(filtered_df.shape)





```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,5))
c= df.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(df['HP'], df['Price'], label='Data Points')  # Scatter plot of HP vs. Price

# Calculate the coefficients (slope and intercept) of the linear regression line
coefficients = np.polyfit(df['HP'], df['Price'], 1)
slope = coefficients[0]
intercept = coefficients[1]

# Create the regression line using the coefficients
regression_line = slope * df['HP'] + intercept

# Plot the regression line
ax.plot(df['HP'], regression_line, color='red', label='Trendline')

ax.set_xlabel('HP')
ax.set_ylabel('Price')
ax.legend()  # Add a legend to the plot
plt.show()
```
